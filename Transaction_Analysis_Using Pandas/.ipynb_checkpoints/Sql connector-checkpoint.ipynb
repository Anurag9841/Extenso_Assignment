```python
## created a function which takes host, user name ,password and database name.
```

```python
def create_connection(host,user,password,database):
    try:
        mydb = mysql.connector.connect(
            host = host,
            user = user,
            password = password,
            database = database
        )
        print("Connction Made sucessfully")
    except error as err:
        print(f"The error is {err}")
    return mydb
```

```python
connection = create_connection("localhost","root","root","client_rw")
```

```python
cursor = connection.cursor()
```

```python
'''
on calling directly shows the tables associated with database
'''
def showtables():
    query = "show tables"
    cursor.execute(query)
    return cursor.fetchall()
```

```python
showtables()
```

```python
'''
on calling directly shows the result of particular table
'''
def select_from_tables(table_name):
    query = f"select * from {table_name}"
    cursor.execute(query)
    return cursor.fetchall()
```

```python
select_from_tables('trans_summ')
```
